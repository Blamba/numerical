Chocolatey
----------
Chocolatey это менеджер пакетов, который позволяет быстро и просто устанавливать нужные нам программы. Как appstore для iphone и googleplay для android :-). Основной идея — не надо искать установщик в интернете, скачивать его и много раз нажимать next; достаточно запустить команду '''choco install _название-пакета_''' и программа будет установлена. Список доступных пакетов можно посмотреть на сайте проекта: https://chocolatey.org/
Итак, для установки надо открыть меню '''Пуск''' и в поиске написать powershell. Открывшаеся окно это обновленная командная оболочка (консоль, shell, терминал) появившаяся в Windows 7 для замены устаревшей консоли CMD. В неё копируем следующую строку и запускаем:

> iex ((new-object net.webclient).DownloadString('https://chocolatey.org/install.ps1'))

Эта команда скачает и запустит установку самого Chocolatey. По окончанию надо будет перезапустить powershell.
Если вы работаете на mac os, то аналог это brew, debian-подобных линуксов это apt-get.

Редактор
--------
Предлагаю пользоваться редактором Atom, это современный редактор с широкими возможностями по расширению функционала. Установить можно скачав с сайта https://atom.io или через choco:

> choco install atom

Основы работы с редактором описаны на https://atom.io/docs/latest/getting-started

Нам нужно поставить плагин для работы с фортраном, для этого заходим в настройки Ctrl-, , выбираем Packages, ищем пакет fortran, нажимаем install.


Компилятор
----------

Watcom fortran, который стоит на компбютерах в классе поддерживает только фортран-77. Чтобы иметь возможность компилировать код на фортран-90 поставим пакет MinGW, который включает в себя gfortran:

> choco install mingw

Компилятор работает из консоли и по-умолчанию называет полученную программу a.exe

> gfortran hello.f90
> a.exe
